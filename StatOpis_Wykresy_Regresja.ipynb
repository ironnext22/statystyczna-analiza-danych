### Zbiór testowy i treningowy

```r
nrow(data)
set.seed(1738)

sample=sample.int(n=nrow(data),size=round(.8*nrow(data)),replace = F) # 80% - zbiór treningowy
```

```r
train=data[sample,]
test=data[-sample,]
dim(train)
dim(test)
```

### Model

```r
#MODEL
model<- lm(charges~.,train)
model
```

```r
summary(model)
```

```r
summary(model)$adj.r.squared
```

### Predykcja

```r
prediction <- predict(model,test)
head(prediction)
```

```r
prediction_train <- predict(model,train)
head(prediction_train)
```

```r
prediction_test <- predict(model,test)
head(prediction_test)
```

```r
results_train <- cbind(prediction_train,train$charges)
colnames(results_train) <- c('pred','real')
results_train <- as.data.frame(results_train)
head(results_train)
```

```r
results_test <- cbind(prediction_test,test$charges)
colnames(results_test) <- c('pred','real')
results_test <- as.data.frame(results_test)
head(results_test)
```

### Metryki

```r
# MSE
print('MSE')
mse_train <- mean((results_train$real-results_train$pred)^2)
mse_train
mse_test <- mean((results_test$real-results_test$pred)^2)
mse_test

#RMSE
print('RMSE')
rmse_train <- mse_train^0.5
rmse_train
rmse_test <- mse_test^0.5
rmse_test

# MAE
print('MAE')
mae_train <- sum(abs(results_train$real-results_train$pred))/nrow(results_train)
mae_train
mae_test <- sum(abs(results_test$real-results_test$pred))/nrow(results_test)
mae_test

#R2
print('R2')
SSE_train=sum((results_train$pred-results_train$real)^2)
SST_train=sum((mean(data$charges)-results_train$real)^2)
R2_train <- 1-SSE_train/SST_train 
R2_train
SSE_test=sum((results_test$pred-results_test$real)^2)
SST_test=sum((mean(data$charges)-results_test$real)^2)
R2_test <- 1-SSE_test/SST_test 
R2_test
```
