```python
plt.hist(df["charges"],bins=10)
plt.ylabel("count")
plt.xlabel("value of charges")
```

```python
fig, ax = plt.subplots(figsize=(16,7))
ax.boxplot(df["charges"])
```

```python
plt.hist(df.loc[df["female"]==1,"charges"],alpha=0.5,label="female")
plt.hist(df.loc[df["female"]==0,"charges"],alpha=0.5,label="male")
plt.legend()
```

```python
fig, ax2 = plt.subplots(figsize=(16,7))
ax2.boxplot([df.loc[df["female"]==0,"charges"],df.loc[df["female"]==1,"charges"]])
```

```python
plt.hist(df.loc[df["smoker"]==1,"charges"],alpha=0.5,label="smoker")
plt.hist(df.loc[df["smoker"]==0,"charges"],alpha=0.5,label="no smoker")
plt.legend()
```

```python
fig, ax2 = plt.subplots(figsize=(16,7))
ax2.boxplot([df.loc[df["smoker"]==0,"charges"],df.loc[df["smoker"]==1,"charges"]])
```

```python
ax = sb.countplot(df["smoker"])
ax.legend(label=["no smoaker","smoaker"])
```
